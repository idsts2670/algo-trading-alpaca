# Alpaca Paper Trading Fundamentals

This notebook demonstrates the core concepts of paper trading with Alpaca's Trading API:
1. Connecting to the Alpaca API
2. Verifying your account and credentials
3. Placing market orders
4. Viewing your portfolio
5. Managing positions

**Note:** This is paper trading (simulated), not real money. Perfect for testing strategies risk-free.

## Step 1: Setup and Configuration

First, we'll import the necessary libraries and load your API credentials.

```python
!uv pip install alpaca-py python-dotenv pytz
```

```python
import os
import sys
from datetime import datetime

from dotenv import load_dotenv
from alpaca.trading.client import TradingClient
from alpaca.trading.requests import MarketOrderRequest, LimitOrderRequest
from alpaca.trading.enums import OrderSide, TimeInForce

# Load environment variables from .env file
load_dotenv()
```

```python
if "google.colab" in sys.modules:
    # In Google Colab environment, we will fetch API keys from Secrets.
    # Please set ALPACA_API_KEY, ALPACA_SECRET_KEY in Google Colab's Secrets from the left sidebar
    from google.colab import userdata

    ALPACA_API_KEY = userdata.get("ALPACA_API_KEY")
    ALPACA_SECRET_KEY = userdata.get("ALPACA_SECRET_KEY")
    
else:
    # Please safely store your API keys and never commit them to the repository (use .gitignore)
    # Load environment variables from environment file (e.g., .env)
    load_dotenv()
    # API credentials for Alpaca's Trading API
    ALPACA_API_KEY = os.environ.get("ALPACA_API_KEY")
    ALPACA_SECRET_KEY = os.environ.get("ALPACA_SECRET_KEY")

## We use paper environment for this example
ALPACA_PAPER_TRADE = True
```

## Step 2: Initialize the Alpaca Trading Client

Create a connection to Alpaca's API using your credentials.

```python
# Initialize the Alpaca trading client
trading_client = TradingClient(api_key=ALPACA_API_KEY, secret_key=ALPACA_SECRET_KEY, paper=ALPACA_PAPER_TRADE)
```

## Step 3: Verify Your Account Connection

Fetch and display your account information to confirm the API connection is working.

```python
# Get account information
account = trading_client.get_account()

# Display key account details
print("Account Information:")
print(f"  Account Status: {account.status}")
print(f"  Account Number: {account.account_number}")
print(f"  Cash: ${float(account.cash):,.2f}")
print(f"  Buying Power: ${float(account.buying_power):,.2f}")
print(f"  Portfolio Value: ${float(account.portfolio_value):,.2f}")
print(f"  Day Trade Count: {account.daytrade_count}")
print(f"  Created at: {account.created_at}")
```

## Step 4: Place a Market Order

Submit a buy order for a stock using a market order (executes at the next available price).

```python
# Define market order parameters
# Buy 1 share of Apple (AAPL)
order_request = MarketOrderRequest(
    symbol="AAPL",
    qty=1,
    side=OrderSide.BUY,
    time_in_force=TimeInForce.DAY
)

# Submit the order
order = trading_client.submit_order(order_data=order_request)

print(f"Order Submitted:")
print(f"  Order ID: {order.id}")
print(f"  Symbol: {order.symbol}")
print(f"  Side: {order.side}")
print(f"  Quantity: {order.qty}")
print(f"  Status: {order.status}")
print(f"  Type: {order.order_type}")
print(f"  Submitted at: {order.submitted_at}")
```

## Step 5: View Your Positions

Retrieve all open positions in your portfolio.

```python
# Get all open positions
positions = trading_client.get_all_positions()

print(f"Open Positions ({len(positions)} total):")
print()

for position in positions:
    print(f"  {position.symbol}")
    print(f"    Quantity: {position.qty}")
    print(f"    Current Price: ${float(position.current_price):,.2f}")
    print(f"    Market Value: ${float(position.market_value):,.2f}")
    print(f"    Unrealized Gain/Loss: ${float(position.unrealized_pl):,.2f} ({float(position.unrealized_plpc)*100:.2f}%)")
    print()
```

## Step 6: Place a Limit Order

Submit a limit order to buy at a specific price.

```python
# Define limit order parameters
# Buy 1 share of Microsoft (MSFT) at $300 or less
limit_order_request = LimitOrderRequest(
    symbol="MSFT",
    qty=1,
    side=OrderSide.BUY,
    limit_price=300.00,
    time_in_force=TimeInForce.GTC  # Good-Till-Canceled
)

# Submit the limit order
limit_order = trading_client.submit_order(order_data=limit_order_request)

print(f"Limit Order Submitted:")
print(f"  Order ID: {limit_order.id}")
print(f"  Symbol: {limit_order.symbol}")
print(f"  Quantity: {limit_order.qty}")
print(f"  Limit Price: ${limit_order.limit_price}")
print(f"  Status: {limit_order.status}")
print(f"  Time in Force: {limit_order.time_in_force}")
```

## Step 7: View Open Orders

Check the status of all open orders.

```python
# Get all open orders
orders = trading_client.get_orders()

print(f"Open Orders ({len(orders)} total):")
print()

for order in orders:
    print(f"  {order.symbol} - {order.side.name}")
    print(f"    Order ID: {order.id}")
    print(f"    Quantity: {order.qty}")
    if order.limit_price:
        print(f"    Limit Price: ${order.limit_price}")
    print(f"    Status: {order.status}")
    print(f"    Submitted: {order.submitted_at}")
    print()
```

## Step 8: Close a Position

Sell a position to close it out.

```python
# Close the AAPL position by selling all shares
close_order_request = MarketOrderRequest(
    symbol="AAPL",
    qty=1,
    side=OrderSide.SELL,
    time_in_force=TimeInForce.DAY
)

close_order = trading_client.submit_order(order_data=close_order_request)

print(f"Close Order Submitted:")
print(f"  Order ID: {close_order.id}")
print(f"  Symbol: {close_order.symbol}")
print(f"  Side: {close_order.side}")
print(f"  Quantity: {close_order.qty}")
print(f"  Status: {close_order.status}")
```

## Step 9: Cancel an Order

Cancel an open order (useful for limit orders that haven't filled).

```python
# Get open orders to find one to cancel
open_orders = trading_client.get_orders()

if open_orders:
    # Cancel the first open order
    order_to_cancel = open_orders[0]
    trading_client.cancel_order(order_id=order_to_cancel.id)
    print(f"Order Canceled:")
    print(f"  Order ID: {order_to_cancel.id}")
    print(f"  Symbol: {order_to_cancel.symbol}")
else:
    print("No open orders to cancel.")
```

## Step 10: Final Portfolio Summary

Get a complete view of your portfolio at this moment.

```python
# Get updated account information
final_account = trading_client.get_account()
final_positions = trading_client.get_all_positions()

print("\n" + "="*50)
print("PORTFOLIO SUMMARY")
print("="*50)
print(f"Portfolio Value: ${float(final_account.portfolio_value):,.2f}")
print(f"Cash: ${float(final_account.cash):,.2f}")
print(f"Buying Power: ${float(final_account.buying_power):,.2f}")
print(f"Total P&L: ${float(final_account.portfolio_value) - 100000:,.2f}")
print(f"Open Positions: {len(final_positions)}")
print("="*50)
```
